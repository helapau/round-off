```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def euler(F,T,Y0,h):
    '''
    F .... 
    T... list/array [t_0,t_f]
    Y0 ... list/array
    '''
    t = np.arange(T[0],T[1],h)
    if t[-1] != T[1]:
        t = np.append(t,T[1])
    t = np.float32(t)
        
    N = np.shape(t)[0]
    Y = np.zeros(shape=(N,1))
    Y[0,:] = Y0
    Y = np.float32(Y)
    
    for i in range(0,N-1):
        Y[i+1,:] = np.float32(Y[i,:] + h*F(t[i],Y[i,:]))
    out = {'t':t,'Y':Y}
    return out
    
    
```

```python
def F(t,X):
    return np.float32(X[0]/t - (X[0]/t)**2)
```

```python
def exa(x):
    return np.float32(x/(1+np.log(x)))
```

```python
T = [1, 2]
Y0 = 1;
h = 0.01;
out = euler(F,T,Y0,h);
plt.plot(out['t'],out['Y'])
plt.xlabel('t')
plt.ylabel('y')
plt.title('approximate w/step '+str(h),fontsize=15)
plt.show()
```

```python
M = 20
er = np.zeros(shape=(M,1))
steps = np.zeros(shape=(M,1))

for j in range(0,M):
    steps[j] = np.float32(1/(2**(j+1)))
    out = euler(F,T,Y0,steps[j])
    er[j] = abs(out['Y'][-1]-exa(2))
```

```python
plt.plot(-np.log2(steps),er)
plt.xlabel('-log2(h)')
plt.ylabel('chyba')
plt.title('celkova chyba',fontsize=15)
plt.show()
```
